```python
from qiskit import *
from qiskit.tools.visualization import *
```

```python
qr = QuantumRegister(2)
cr = ClassicalRegister(2)
qc_1 = QuantumCircuit(qr,cr)
qc_1.draw(output = 'mpl')
```

```python
qc_1.x(0)
qc_1.h(1)
qc_1.barrier()
qc_1.draw(output = 'mpl')
```

```python
qc_2 = QuantumCircuit(qr,cr)
qc_2.h(0)
qc_2.x(1)
qc_2.barrier()
qc_2.draw(output = 'mpl')
```

```python
qc_3 = QuantumCircuit(qr, cr)
qc_3.h(0)
qc_3.x(1)
qc_3.measure(0,1)
qc_3.measure(1,0)
qc_3.draw(output='mpl')
```

```python
qc_4 = qc_1.compose(qc_2)
qc_4 = qc_4.compose(qc_3)
qc_4.draw(output='mpl')
```
