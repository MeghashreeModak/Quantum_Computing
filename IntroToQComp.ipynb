# STEPS TO CREATE THE SIMPLEST CIRCUIT USING QISKIT (without any gates)

```python
# step1: import qiskit
from qiskit import *
from qiskit.tools.visualization import * # for result drawing in the next cell

# step2: create classical and quantum registers
qr = QuantumRegister(2)
cr = ClassicalRegister(2)

# step3: Create a QuantumCircuit object
circuit = QuantumCircuit(qr, cr)

# step4: draw the circuit in ASCII
# circuit.draw()

# To draw the circuit in picture
"""
    circuit.draw(output='mpl') may throw an error regarding an optional library,
    in that case run `pip install pylatexenc`
"""
circuit.draw(output='mpl')
```

# CREATING QUANTUM HALF ADDER

[For resource](https://qiskit.org/textbook/ch-states/atoms-computation.html)

```python
circuit.x(0)
circuit.x(1)
circuit.cx(0, 1)

circuit.measure(0, 0)
circuit.measure(1, 1)

circuit.draw(output='mpl')
```

```python
qc_ha = QuantumCircuit(4,2)
# encode inputs in qubits 0 and 1
qc_ha.x(0) # For a=0, remove this line. For a=1, leave it.
qc_ha.x(1) # For b=0, remove this line. For b=1, leave it.
qc_ha.barrier()
# use cnots to write the XOR of the inputs on qubit 2
qc_ha.cx(0,2)
qc_ha.cx(1,2)
qc_ha.barrier()
# extract outputs
qc_ha.measure(2,0) # extract XOR value
qc_ha.measure(3,1)

qc_ha.draw(output='mpl')
```

```python
sim = Aer.get_backend('aer_simulator') 
result = sim.run(qc_ha).result()
counts = result.get_counts()
plot_histogram(counts)
```
